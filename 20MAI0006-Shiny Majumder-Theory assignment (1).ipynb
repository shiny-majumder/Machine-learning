So, in one tuple i have shown no. of people using lipstick after crossing expire date and if this is creating any side effects or not. I have used 10 tuples here based on my choice. </b>





<b>Firstly I am using Logistic Regression Classifier.</b>

```python
#1 import the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<b> Screenshot of Dataset:</b>


```python
#2 Import and read the dataset

train=pd.read_csv('F:/VIT COLLEGE/2nd SEM/MACHINE LEARNING/LAB/Lipstick_dataset.csv') #lipstick_dataset.csv
train.head(15)
```

```python
#3 fix the predictor

df=train[['No of people used after Expire date','Side Effects of using lipstick after expire date' ]]
```

```python
#4 take a look at the dataframe

df.head(15)
```

```python
#5 set the predictor(X) and Response(Y) variables

X= df.drop("Side Effects of using lipstick after expire date", axis=1)
Y= df["Side Effects of using lipstick after expire date"]
```

```python
#6 Splitting into the training and test set

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=25)
```

```python
#7 call the logisctic regression model

from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(X_train,Y_train)
```

```python
#8 Compute the predictions or Y hat

Y_pred =logit.predict(X_test)
```

```python
#11 confusion matrix

from sklearn.metrics import confusion_matrix
confusion_matrix= confusion_matrix(Y_test,Y_pred)
confusion_matrix
```

```python
#9 accuracy score

from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```

```python
#13 classification report

from sklearn.metrics import classification_report
report = classification_report(Y_test, Y_pred)
print(report)
```

<b>Using Naive Bayes Classifer</b>

```python
#1 call the naive bayes model

from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
Y_pred = gnb.fit(X_train, Y_train).predict(X_test)
```

```python
#2 accuracy score

from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```

<b>Conclusion:</b> From this dataset , I can conclude that for logistic regression classifier the accuracy is higher (0.8) where the accuracy for naive bayes classifier is lower (0.2)
