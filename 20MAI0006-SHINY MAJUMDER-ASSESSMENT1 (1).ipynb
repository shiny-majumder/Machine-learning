```python
#Task 6: Implementation of Five Feature Engineering Task
    
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Square_area'] = df['Square_area'].astype(int)
```

```python
sns.factorplot(x = "Square_area", y = 'Price', data = df)
```

```python
#Task 5: Plotting  bar graph betweeen 'Place' and 'price'
groupedby_places = df.groupby('Location')[['Price']].mean()
print(groupedby_places.sort_values(by = ['Price']))
plt.figure(figsize=(9,6))
plt.bar(x=df['Location'],height=df['Price'])
plt.xticks(rotation=45)
```
