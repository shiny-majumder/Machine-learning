```python
#converting into dictionary
data['month']
data['month'].value_counts()
month_issued_dict=data['month'].value_counts().to_dict()
month_issued_dict
```

```python
#function to return key for any value in a dictionary

def get_key(val):
    for key,value in month_issued_dict.items():
        if val == value:
            return key
    return "key dosen't exist"
max_issues=max(data['month'].value_counts())
print("The month with maximum number of loan issued", get_key(max_issues))
```
